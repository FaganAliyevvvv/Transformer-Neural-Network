```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

**Model 1**

```python
model1 = SentenceTransformer('intfloat/multilingual-e5-large')
```

```python
wordsforembeddings = [
    'memar',
    'yaradıcı memar',
    'təcrübəli memar'
]
```

```python
embeddings1 = model1.encode(wordsforembeddings)
```

```python
cos_sim1 = cosine_similarity(embeddings1)
```

```python
sns.heatmap(cos_sim1, annot=True, cmap="YlGnBu",xticklabels=wordsforembeddings, yticklabels=wordsforembeddings)
```

# Fine Tuning
--------------------------------------------------------------------------------------------------------------------------------------

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
pip install datasets
```

```python
from datasets import load_dataset

dataset_id = "embedding-data/QQP_triplets"
dataset = load_dataset(dataset_id)
```

```python
from sentence_transformers import InputExample

train_examples = []
train_data = dataset['train']['set']
n_examples = dataset['train'].num_rows // 2

for i in range(n_examples):
  example = train_data[i]
  train_examples.append(InputExample(texts=[example['query'], example['pos'][0], example['neg'][0]]))
```

```python
from torch.utils.data import DataLoader
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=16)
```

```python
from sentence_transformers import losses

train_loss = losses.TripletLoss(model=model)
```

```python
model.fit(train_objectives=[(train_dataloader, train_loss)], epochs=10)
```

```python
tunnedembeddings = model.encode(englishwords)
```

```python
tunnedcos_sim = cosine_similarity(tunnedembeddings)
```

# ---------------------------------------------------------------------------------------------------------------

```python
sns.heatmap(tunnedcos_sim, annot=True, cmap="YlGnBu",xticklabels=englishwords, yticklabels=englishwords)
```

**Before VS After**

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 6), gridspec_kw={'wspace': 0.3})

ax[0].set_title("Before")
sns.heatmap(engcos_sim, annot=True, cmap="YlGnBu", xticklabels=englishwords, yticklabels=englishwords, ax=ax[0])

ax[1].set_title("After")
sns.heatmap(tunnedcos_sim, annot=True, cmap="YlGnBu", xticklabels=englishwords, yticklabels=englishwords, ax=ax[1])

plt.show()
```

**Fine Tunned English language model vs Azerbaijan language model**

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 6), gridspec_kw={'wspace': 0.3})

ax[0].set_title("English")
sns.heatmap(tunnedcos_sim, annot=True, cmap="YlGnBu", xticklabels=englishwords, yticklabels=englishwords, ax=ax[0])

ax[1].set_title("Azerbaijan")
sns.heatmap(cos_sim1, annot=True, cmap="YlGnBu", xticklabels=wordsforembeddings, yticklabels=wordsforembeddings, ax=ax[1])

plt.show()
```
